```python
import pandas as pd # Imports the Python data analysis library "pandas".

df = pd.read_csv("courses.csv") # Reads the csv file created w/ all relevant information on UMD classes into a dataframe.
display(df) # Displays the dataframe - for debugging purposes, removed after final iteration of program.
```

```python
# Coded by: Ady Weng, Michael Pulley, Lorenzo Regala

def sort_courses(): # Intended to sort courses by the specified user input criteria.
    
    # Asks for and identifies the user input criteria.
    
    print("Please choose to sort by one of the following:")
    key_word = input("Course Number, Name, Avg. Semester Grade, Pass Rate, Professor, Professor Rating, or Format")
    
    # Code is expected to take the input "key_word" and compare it to string values of criteria that the data can be sorted by;
    ## If they match, then another dataframe will subsequently be created with newly sorted rows of data using the .sort_values function and displayed.
    
    if key_word == "Course Number":
        co_num_df = df.sort_values(by='Course_#')
        display(co_num_df)
    elif key_word == "Name":
        name_df = df.sort_values(by='Name')
        display(name_df)
    elif key_word == "Avg. Semester Grade":
        grade_df = df.sort_values(by='Average_Semester_Grade')
        display(grade_df)
    elif key_word == "Pass Rate":
        pass_df = df.sort_values(by='Pass_Rate')
        display(pass_df)
    elif key_word == "Professor":
        prof_df = df.sort_values(by='Professor')
        display(prof_df)
    elif key_word == "Professor Rating":
        prof_rate_df = df.sort_values(by='Professor_Rating')
        display(prof_rate_df)
    elif key_word == "Format":
        form_df = df.sort_values(by='Format')
        display(form_df)
    else:
        print("Try again!")
        
    # Still need to implement the try-catch loops that will negate incorrect syntax for inputs;
    ## As of now, the inputs must be exactly as they are written in the instructions "Please choose to sort by one of the following: 'Course Number', 'Name,' etc."
    ### Hopefully, the try-catch loops will enable us to redirect the user to input a key_word correctly instead of breaking and freezing the program.
```

```python
# Coded by: Michael Pulley, Tristan Clark, Andrew Birgin

def search_courses(): # Intended to search for one or more course that specifically pertains to the user's search criteria.
    
    # Asks for and identifies which category the user would like to search by, and what "search_term" should be used;
    ## e.g., Selecting "Name" and then entering "Data Organization" will return all classes with that name.
    
    print("Please choose to search one of the following:")
    key_word = input("Course Number, Name, Avg. Semester Grade, Pass Rate, Professor, or Format")
    search_term = input("Now, enter your search criteria:")
    
    # Code is expected to take the input "key_word" and compare it to string values of criteria that the data can be sorted by;
    ## Then, the user's "search_term" criteria is compared to all values within that specific category in the dataframe;
    ### (For example, the if the key_word is "Course Number" and the search_term entered is "311", then "311" will be searched for amongst all course numbers.)
    #### Subsequently, each row that satisfies the new criteria are read into a new dataframe and displayed.
    
    if key_word == "Course Number":
        co_num_df = df[df['Course_#'] == int(search_term)]
        display(co_num_df)
    elif key_word == "Name":
        name_df = df[df['Name'] == search_term]
        display(name_df)
    elif key_word == "Avg. Semester Grade":
        try: # Try-except to ensure that there is no ValueError vis-Ã -vis converting the String input 'search_term' into an integer;
            grade_df = df[df['Average_Semester_Grade'] == int(search_term)]
            display(grade_df)
        except: # Alerts the user to the fact that a numerical search term must be used, allows them another opportunity.
            print("Oops! An error occurred; try a numerical search term.")
            search_term = input("Now, enter your search criteria:")
            grade_df = df[df['Average_Semester_Grade'] == int(search_term)]
            display(grade_df)
    elif key_word == "Pass Rate":
        pass_df = df[df['Pass_Rate'] == int(search_term)]
        display(pass_df)
        try: # Try-except w/ same intention as aforementioned;
            pass_df = df[df['Pass_Rate'] == int(search_term)]
            display(pass_df)
        except: # Similarly allows the user to make changes to their input to fit with the necessary typing.
            print("Oops! An error occurred; try a numerical search term.")
            search_term = input("Now, enter your search criteria:")
            pass_df = df[df['Pass_Rate'] == int(search_term)]
            display(pass_df)
    elif key_word == "Professor":
        prof_df = df[df['Professor'] == search_term]
        display(prof_df)
    elif key_word == "Format":
        form_df = df[df['Format'] == search_term]
        display(form_df)
    else:
        print("Try again!")        
        
    # Similar to sort_courses; still need to implement the try-catch loops that will negate incorrect syntax for inputs;
    ## As of now, the inputs must be exactly as they are written in the instructions "Please choose to sort by one of the following: 'Course Number', 'Name,' etc."
    ### Unlike the sort_courses function, the input for the search_term does not have to be exact to continue working;
    #### For example, the program will return "Data Organization"... "Data Science"... for the query "Data".
```

```python
# Coded by: Ady Weng, Michael Pulley, Lorenzo Regala

def save_courses(save_option): # Lets the user save the courses based on their row number (leftmost value); accepts input of "save_option", which dictates whether or not the user would like to save.
    
    saving = True # Initializes variable to determine whether or not the code is still saving - allows the user to save more than one course in a while-loop.
    
    if save_option == "Yes": # If the user first wanted to save a course;
        while saving == True: # Ensures that saving keeps running unless the user says no, upon which saving == False and this condition is negated;
            
            # Asks user for the row number of the course they would like to save and saves it within the global list "saved_courses".
            
            num = input("Enter the leftmost value of the course you would like to save.")
            saved_courses.append(int(num))
            
            # Asks user if they would like to continue - runs as normal until the user ansewrs no, which is when the boolean "saving" flips to False.
            
            save_option = input("Continue? Yes or No.")
            if save_option == "No":
                saving = False
```

```python
# Coded by: Ady Weng

def function_select(option): # Allows the user to determine which function (sorting or selecting) they would like to utilize; also allows the program to continuously run until manually quit.
    
    # Initialization of booleans "running" and "not_First".
    
    running = True
    not_First = True
    
    while running == True: # While the user continues to want the program to run;
        
        # Ensures there is no repeat of the "option" input in the below cell, as "Would you like to sort or search?" would be repeated twice on the program's first run.
        
        if not_First == False:
            option = input("Would you like to sort or search?")
            
        # Determines which function is called by the user_input "option" first determined in the below cell and taken as an argument for this function.
            
        if option == "Sort":
            sort_courses()
        elif option == "Search":
            search_courses()
        else:
            print("Invalid Option")
            
        # Offers to save the course - subsequently runs the "save_courses" function with their choice (if No, nothing occurs, and the program continues running).
            
        save_option = input("Would you like to save a course? Yes or No.")
        save_courses(save_option)
        
        # Asks if the user would like to continue running the program; if they answer no, then the "running" variable is set to False;
        ## Ends the program by finding the saved courses via their row number using the .iloc function, and printing them and their pertinent information in a formatted manner.
            
        cont_running = input("Would you like to continue running the program? Yes or No.")
        if cont_running == "No":
            running = False
            for i in saved_courses:
                print(f"{df.iloc[i]}\n----------------------\n")
                
        not_First = False # Upon first run of this function, the not_First boolean is set to false and this function's "option" input is allowed to run
        ## As the not_First boolean is first initialized outside the loop, it is not set to True again.
```
